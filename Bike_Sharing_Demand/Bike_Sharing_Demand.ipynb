Data prep

```python
dataset_categorical = dataset.select_dtypes(include=['object'])
dataset_categorical.head()
```

```python
dataset_dummies = pd.get_dummies(dataset_categorical, drop_first=True).astype(int)
dataset_dummies.head()
```

Drop catagorical variable columns

```python
dataset = dataset.drop(list(dataset_categorical),axis=1)
dataset
```

```python
#concat dummies with the dataset
dataset = pd.concat([dataset,dataset_dummies],axis=1)
dataset.head()
```

```python
dataset = dataset.drop(['instant','dteday'],axis=1,inplace=False)
dataset.head()
```

Model Building and eval

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
np.random.seed(0)
```

```python
# splitting in train and test sets

df_train,df_test= train_test_split(dataset,train_size=0.7,test_size=0.3,random_state=100)
df_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
#applying scaler to all except dummy
var = ['temp','hum','windspeed','casual','registered','cnt']
df_train[var] =scaler.fit_transform(df_train[var])
df_train
```

```python
df_train.describe()
```

```python
# checking cor coeff
plt.figure(figsize=(30,30))
sns.heatmap(df_train.corr(),annot=True,cmap='YlGnBu')
plt.show()
```

```python
# Dividing into x and y

y_train = df_train.pop('cnt')
x_train = df_train.drop(['casual','registered'],axis=1)

x_train.head()
```

```python
np.array(x_train)
```

```python
import statsmodels.api as sm
x_train_lm = sm.add_constant(x_train)

lr = sm.OLS(y_train,x_train_lm).fit()
lr.params
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
lr.summary()
```

```python
from sklearn.feature_selection import RFE
```

```python
lm = LinearRegression()
rfe1 = RFE(estimator=lm, n_features_to_select=15)

# fitting with 15 features

rfe1.fit(x_train,y_train)
print(rfe1.support_)
print(rfe1.ranking_)
```

```python
col1 = x_train.columns[rfe1.support_]
col1
```

```python
x_train_rfe1 = x_train[col1]
x_train_rfe1 = sm.add_constant(x_train_rfe1)
lm1 = sm.OLS(y_train,x_train_rfe1).fit()
lm1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

a = x_train_rfe1.drop('const',axis=1)
```

```python
#eval VIF

vif = pd.DataFrame()
vif['features']=a.columns
vif['VIF']= [variance_inflation_factor(a.values,i) for i in range(a.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF' , ascending= False)
vif
```

```python
lm = LinearRegression()
rfe2 = RFE(estimator=lm, n_features_to_select=7)

# fitting with 15 features

rfe2.fit(x_train,y_train)
print(rfe2.support_)
print(rfe2.ranking_)
```

```python
col2 = x_train.columns[rfe2.support_]
col2
```

```python
x_train_rfe2 = x_train[col2]
x_train_rfe2 = sm.add_constant(x_train_rfe2)
lm2 = sm.OLS(y_train,x_train_rfe2).fit()
lm2.summary()
```

```python
#eval VIF
a1 = x_train_rfe2.drop('const',axis=1)
vif1 = pd.DataFrame()
vif1['features']=a1.columns
vif1['VIF']= [variance_inflation_factor(a1.values,i) for i in range(a1.shape[1])]
vif1['VIF'] = round(vif['VIF'],2)
vif1 = vif1.sort_values(by='VIF' , ascending= False)
vif1
```

```python
y_train_cnt = lm2.predict(x_train_rfe2)
ffig = plt.figure()
sns.distplot((y_train,y_train_cnt),bins=20)
plt.show()
```

```python
# 📊 Model Performance Evaluation for Bike Sharing Demand Prediction
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming these variables exist from your code:
# y_test, y_pred

# Calculate metrics
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

# Print neatly formatted results
print("🚴‍♂️ Bike Sharing Demand Model Performance 🚴‍♀️")
print(f"R² Score  : {r2:.4f}")
print(f"RMSE      : {rmse:.4f}")
print(f"MAE       : {mae:.4f}")

# Visualize Actual vs Predicted
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred, color='orange', s=60, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Actual vs Predicted Bike Demand', fontsize=14)
plt.xlabel('Actual Demand')
plt.ylabel('Predicted Demand')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Visualize residuals
residuals = y_test - y_pred
plt.figure(figsize=(8,6))
sns.histplot(residuals, bins=25, kde=True, color='teal')
plt.title('Residual Distribution', fontsize=14)
plt.xlabel('Prediction Error (Residuals)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Summary bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=['R² Score', 'RMSE', 'MAE'], y=[r2, rmse, mae], palette='magma')
plt.title('Model Performance Metrics', fontsize=14)
plt.ylabel('Score / Error')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```
